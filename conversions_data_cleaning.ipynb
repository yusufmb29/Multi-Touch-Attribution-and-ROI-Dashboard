## Conversions Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Yusuf\OneDrive\Desktop\conversions.csv")
df
```

```python
print(df.columns.tolist())
```

```python
# Strip spaces from all string columns
for col in df.select_dtypes(include='object'):
    df[col] = df[col].str.strip()
```

```python
#checking duplicates
print(df.duplicated().sum())
```
