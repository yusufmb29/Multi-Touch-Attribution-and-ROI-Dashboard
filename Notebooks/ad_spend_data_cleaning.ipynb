## Ad Spent Data Cleaning 

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Yusuf\OneDrive\Desktop\ad_spend.csv")
df
```

```python
print(df.columns.tolist())
```

```python
df.drop(columns=['spend_id', 'ad_group_id','cpc', 'ctr'], inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
# Strip spaces from all string columns
for col in df.select_dtypes(include='object'):
    df[col] = df[col].str.strip()
```

```python
#checking duplicates
print(df.duplicated().sum())
```
